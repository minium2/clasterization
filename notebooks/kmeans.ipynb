```python
from sklearn.datasets import make_blobs
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from scripts.kmeans import k_means
from scripts.dbscan import dbscan

from sklearn.cluster import DBSCAN
```

```python
dataset, classes = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=0.5, random_state=0)
# make as panda dataframe for easy understanding
df = pd.DataFrame(dataset, columns=['var1', 'var2'])
df.head(2)
```

```python

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df)
sns.scatterplot(data=df, x="var1", y="var2", hue=kmeans.labels_)
plt.show()
```

```python
custom_k_means = k_means(dataset, k=4)
labels = [a for a, _ in custom_k_means]
sns.scatterplot(data=df, x="var1", y="var2", hue=labels)
plt.show()
```

```python

custom_dbscan = dbscan(m=dataset.transpose(), eps=0.4, min_points=5)
sns.scatterplot(data=df, x="var1", y="var2", hue=custom_dbscan)
plt.show()
```

```python
dbscan = DBSCAN()
dbscan.fit(dataset)

sns.scatterplot(data=df, x="var1", y="var2", hue=dbscan.labels_)
plt.show()
```
